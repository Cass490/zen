```python
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt
from tqdm import tqdm
from sklearn.metrics import mean_squared_error

# --- Step 1: Install PennyLane if you haven't ---
# In your terminal: pip install pennylane pennylane-lightning
import pennylane as qml

# --- Configuration ---
WINDOW_SIZE = 10
N_QUBITS = 4
HIDDEN_SIZE = 16
EPOCHS = 50
LEARNING_RATE = 0.005
BATCH_SIZE = 32

# --- Device Setup ---
QUANTUM_DEVICE_NAME = "lightning.qubit"
TORCH_DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")

print(f"Using PyTorch device: {TORCH_DEVICE}")
print(f"Using PennyLane device: {QUANTUM_DEVICE_NAME}")


# --- Data Loading and Preparation ---
def load_and_prepare_data(path="laser_train.csv", window=WINDOW_SIZE):
    try:
        df = pd.read_csv(path, header=None, names=["laser"])
        data = df["laser"].values
    except FileNotFoundError:
        print(f"Error: Could not find the dataset at '{path}'.")
        print("Please make sure 'laser_train.csv' is in the same directory.")
        return None, None, None, None

    data_mean = np.mean(data)
    data_std = np.std(data)
    data_normalized = (data - data_mean) / data_std

    X, y = [], []
    for i in range(len(data_normalized) - window):
        X.append(data_normalized[i:i + window])
        y.append(data_normalized[i + window])
    
    X, y = np.array(X), np.array(y)

    train_size = int(0.7 * len(X))
    X_train, y_train = X[:train_size], y[:train_size]
    X_test, y_test = X[train_size:], y[train_size:]

    X_train = torch.FloatTensor(X_train).unsqueeze(-1).to(TORCH_DEVICE)
    y_train = torch.FloatTensor(y_train).to(TORCH_DEVICE)
    X_test = torch.FloatTensor(X_test).unsqueeze(-1).to(TORCH_DEVICE)
    y_test = torch.FloatTensor(y_test).to(TORCH_DEVICE)

    return X_train, y_train, X_test, y_test


# --- Quantum Circuit and Model Definition ---

@qml.qnode(qml.device(QUANTUM_DEVICE_NAME, wires=N_QUBITS), interface="torch")
def quantum_circuit(inputs, weights):
    qml.AngleEmbedding(inputs, wires=range(N_QUBITS), rotation='Y')
    qml.BasicEntanglerLayers(weights, wires=range(N_QUBITS))
    return qml.expval(qml.PauliZ(0))


class QuantumEnhancedLSTM(nn.Module):
    def __init__(self, input_size, hidden_size, n_qubits):
        super().__init__()
        self.hidden_size = hidden_size
        self.n_qubits = n_qubits
        
        self.lstm = nn.LSTM(
            input_size=input_size,
            hidden_size=hidden_size,
            num_layers=1,
            batch_first=True
        )
        
        n_quantum_layers = 2
        weight_shapes = {"weights": (n_quantum_layers, n_qubits)}
        self.quantum_layer = qml.qnn.TorchLayer(quantum_circuit, weight_shapes)
        self.classical_readout = nn.Linear(1, 1)

    def forward(self, x):
        _, (hidden_state, _) = self.lstm(x)
        hidden_state = hidden_state.squeeze(0)
        
        if self.hidden_size < self.n_qubits:
            padding = torch.zeros(x.shape[0], self.n_qubits - self.hidden_size, device=x.device)
            hidden_state_padded = torch.cat([hidden_state, padding], dim=1)
        else:
            hidden_state_padded = hidden_state[:, :self.n_qubits]
            
        quantum_output = self.quantum_layer(hidden_state_padded)
        quantum_output_reshaped = quantum_output.unsqueeze(-1)
        
        return self.classical_readout(quantum_output_reshaped).squeeze()


# --- Main Experiment Function ---
def run_qlstm_experiment():
    X_train, y_train, X_test, y_test = load_and_prepare_data()
    if X_train is None:
        return

    train_dataset = torch.utils.data.TensorDataset(X_train, y_train)
    train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True)
    
    model = QuantumEnhancedLSTM(
        input_size=1,
        hidden_size=HIDDEN_SIZE, 
        n_qubits=N_QUBITS
    ).to(TORCH_DEVICE)
    
    criterion = nn.MSELoss()
    optimizer = optim.Adam(model.parameters(), lr=LEARNING_RATE)
    
    print("--- Starting QLSTM Training ---")
    
    for epoch in range(EPOCHS):
        model.train()
        running_loss = 0.0
        pbar = tqdm(train_loader, desc=f"Epoch {epoch+1}/{EPOCHS}")
        
        for inputs, labels in pbar:
            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
            running_loss += loss.item()
            pbar.set_postfix({'batch_loss': f'{loss.item():.4f}'})
        
        avg_train_loss = running_loss / len(train_loader)
        
        model.eval()
        with torch.no_grad():
            test_outputs = model(X_test)
            test_loss = criterion(test_outputs, y_test).item()
        
        print(f"Epoch {epoch+1} Summary | Train Loss: {avg_train_loss:.6f} | Test Loss: {test_loss:.6f}")

    print("\n--- Training Complete. Evaluating Final Model... ---")
    model.eval()
    with torch.no_grad():
        final_predictions = model(X_test).cpu().numpy()
    
    ground_truth = y_test.cpu().numpy()
    
    nmse = np.mean((ground_truth - final_predictions)**2) / np.var(ground_truth)
    print(f"Final Test NMSE: {nmse:.6f}")
    
    plt.figure(figsize=(15, 7))
    plt.plot(ground_truth, "k-", label='Ground Truth')
    plt.plot(final_predictions, "r--", label=f'QLSTM Prediction (NMSE: {nmse:.4f})', alpha=0.8)
    plt.title('Quantum-Enhanced LSTM Prediction vs. Ground Truth', fontsize=16)
    plt.xlabel('Time Step (in test set)', fontsize=12)
    plt.ylabel('Normalized Intensity', fontsize=12)
    plt.legend(fontsize=12)
    plt.grid(True)
    plt.savefig("qlstm_efficient_prediction.png")
    plt.show()


if __name__ == "__main__":
    run_qlstm_experiment()
```
